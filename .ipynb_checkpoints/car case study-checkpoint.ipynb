# Checking the OLS assumption

```python
f, (ax1,ax2,ax3) = plt.subplots(1, 3, sharey=True, figsize= (15,3))
ax1.scatter(data_clean['Year'],data_clean["Price"])
ax1.set_title("price and year")
ax2.scatter(data_clean['EngineV'],data_clean["Price"])
ax2.set_title("EngineV and year")
ax3.scatter(data_clean['Mileage'],data_clean["Price"])
ax3.set_title("Mileage and year")
plt.show
```

# relaxing the assumption

```python
log_price = np.log(data_clean["Price"])
data_clean["log_price"] = log_price
data_clean
```

```python
f, (ax1,ax2,ax3) = plt.subplots(1, 3, sharey=True, figsize= (15,3))
ax1.scatter(data_clean['Year'],data_clean["log_price"])
ax1.set_title("price and year")
ax2.scatter(data_clean['EngineV'],data_clean["log_price"])
ax2.set_title("EngineV and year")
ax3.scatter(data_clean['Mileage'],data_clean["log_price"])
ax3.set_title("Mileage and year")
plt.show
```

# creating Dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multico, drop_first=True)
data_with_dummies
```

```python
data_with_dummies.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables =  data_with_dummies[['Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i)for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
```

```python
data_preprocessed
```

# Linear Regression model

# declear the input and targets

```python
targets = data_preprocessed["log_price"]
```

```python
inputs = data_preprocessed.drop(["log_price"],axis=1)
inputs
```

# scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input_scaled,targets, test_size = 0.2, random_state = 365)
```

# create the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Target (y_train)')
plt.ylabel('Predictions (y_hat)')
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
```

```python
reg.score(x_train,y_train)
```

# Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ["Features"])
reg_summary["weights"] = reg.coef_
reg_summary
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel("Targent (y_test)")
plt.ylabel("Prediction (y_hat_test)")
plt.xlim(6,13)
plt.ylim(6,13)
plt.show
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=["Prediciton"])
```

```python
df_pf
```

```python
df_pf["Target"] = np.exp(y_test)
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf
```

```python
df_pf["Residual"] = df_pf["Target"] - df_pf["Prediciton"]
```

```python
df_pf["Difference%"] = np.absolute(df_pf["Residual"]/df_pf["Target"]*100)
```

```python
df_pf
```

```python
df_pf.describe(include ="all")
```
